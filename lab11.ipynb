```python
import numpy as np
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# 1. Laden und Plotten der Audio-Datei
audio_file_path = "Aufzeichnung.wav"
sampling_rate, data = wavfile.read(audio_file_path)
data = np.clip(data, -32768, 32767)  # Ensure data is in the correct range for int16

# Plot the audio signal
plt.figure(figsize=(10, 4))
plt.plot(data)
plt.title("Audio Signal")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xlim(0, len(data))
plt.show()

# 2. Erzeugen und Analysieren eines harmonischen Signals
f0 = 440  # Grundfrequenz
A = 1.0  # Amplitude
t_duration = 1.0  # Dauer in Sekunden
sampling_rate_harmonic = 20 * f0  # Abtastrate

t = np.linspace(0, t_duration, int(sampling_rate_harmonic * t_duration), endpoint=False)
harmonic_signal = A * (np.sin(2 * np.pi * f0 * t) + 
                       np.sin(2 * np.pi * 2 * f0 * t) + 
                       np.sin(2 * np.pi * 3 * f0 * t))

# Plot des harmonischen Signals
plt.figure(figsize=(10, 4))
plt.plot(t, harmonic_signal)
plt.title("Harmonic Signal")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.xlim(0,0.01)
plt.show()

# Speichern des harmonischen Signals als WAV-Datei
wavfile.write("harmonic_signal.wav", sampling_rate_harmonic, harmonic_signal.astype(np.float32))

# 3. Phase der Töne ändern und analysieren
np.random.seed(0)  # Für Reproduzierbarkeit
phases = np.random.uniform(0, 2 * np.pi, 3)
harmonic_signal_phase = A * (np.sin(2 * np.pi * f0 * t + phases[0]) + 
                             np.sin(2 * np.pi * 2 * f0 * t + phases[1]) + 
                             np.sin(2 * np.pi * 3 * f0 * t + phases[2]))

# Plot des phasenmodifizierten harmonischen Signals
plt.figure(figsize=(10, 4))
plt.plot(t, harmonic_signal_phase)
plt.title("Phase-modified Harmonic Signal")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.xlim(0, 0.01)
plt.show()

# 4. Kammerton komplett auslöschen
cancellation_signal = -np.sin(2 * np.pi * f0 * t)
resulting_signal = harmonic_signal + cancellation_signal

# Plot des resultierenden Signals
plt.figure(figsize=(10, 4))
plt.plot(t, resulting_signal)
plt.title("Resulting Signal with Fundamental Frequency Cancelled")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.xlim(0, 0.01)
plt.show()

# 5. Modellierung der periodischen Rechteckfunktion
f0_square = 1
sampling_rate_square = 8000  # Abtastrate für die Rechteckfunktion
t_square = np.linspace(0, 1, int(sampling_rate_square), endpoint=False)

# Generieren der Rechteckfunktion mit den ersten 9 Harmonischen
square_wave = np.zeros_like(t_square)
for n in range(1, 10, 2):
    square_wave += (4 / (np.pi * n)) * np.sin(2 * np.pi * n * f0_square * t_square)

# Plot der Rechteckfunktion
plt.figure(figsize=(10, 4))
plt.plot(t_square, square_wave)
plt.title("Square Wave Signal with First 9 Harmonics")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.xlim(0, 1)
plt.show()

# Generieren und Plotten der 1. und 9. Harmonischen
harmonic_1st = (4 / np.pi) * np.sin(2 * np.pi * f0_square * t_square)
harmonic_9th = (4 / (9 * np.pi)) * np.sin(2 * np.pi * 9 * f0_square * t_square)

plt.figure(figsize=(10, 4))
plt.plot(t_square, harmonic_1st, label="1st Harmonic")
plt.plot(t_square, harmonic_9th, label="9th Harmonic")
plt.legend()
plt.title("1st and 9th Harmonics of the Square Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.xlim(0, 1)
plt.show()
```
